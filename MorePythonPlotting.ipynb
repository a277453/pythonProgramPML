Now consider the following data points that we want to plot for a measurement of some vehicle versus time.
time      0.0 1.0  2.0  3.0
Distance  5.0 9.0 13.0 17.0
How do we plot it? try the following.

```python
time = [0.0, 1.0, 2, 3]
distance = [5.0, 9.0, 13, 17]
plt.plot(time, distance)
plt.xlabel('time')
plt.ylabel('distance');
```

What if we didn't want the lines but want to see just the points?

```python
plt.plot(time, distance, 'o');
```

```python
plt.plot(time, distance, '-.'); # Line styles '-' or '.' or '-.' etc
plt.plot([1.1 * ele for ele in time], [1.1 * dis for dis in distance], '--')
plt.plot([1.2 * ele for ele in time], [1.2 * dis for dis in distance], '.')
plt.plot([1.3 * ele for ele in time], [1.3 * dis for dis in distance], 'o')
plt.legend(['original values', 'Plus 10 %', 'Plus 20 %', 'Plus 30 %'])
plt.xlabel('time')
plt.ylabel('distance');
```

# List slicing
Slicing always produces a list. the general syntax is

list[initial:final:step]

* intial / final are indices and the final is not included.
* negative indices start from the end (-1 is the last element).
* Exmaples make it easier.

```python
lst[0:]
```

```python
lst[:-1]
```

```python
lst[0:4:2]
```

```python
lst[::2]
```

```python
lst[::-1]
```

```python
lst[:200:2]
```
