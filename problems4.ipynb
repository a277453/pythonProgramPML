1. Consider the 7 different batsman data provided to you. Compute the mean,
   median, mode, standard deviation for each of these batsmen. Store the names
   of the batsmen in a list. Then store the mean/median/mode for each batsman
   in the same order.  You should have 5 lists with you at the end of this.

```python
# another way to parse all files in the directory
# folder_path = "..\\data\\"
# file_list = glob.glob(os.path.join(folder_path, "*.txt"))
# print(file_list)

file_list = ['..\\data\\dhoni.txt', '..\\data\\gambhir.txt', '..\\data\\kapil_dev.txt',
             '..\\data\\kohli.txt', '..\\data\\rohit.txt', '..\\data\\sachin.txt', '..\\data\\sehwag.txt']
```

```python
def getDataFromFile(filename):
    with open(filename, "r") as f:
        scores = list(map(int, f.read().strip().split()))
    return scores

def get_frequency_list(lst):
    unq_lst = list(set(lst))
    
    res_val = []
    res_freq = []
    for val in unq_lst:
        res_val.append(val)
        res_freq.append(lst.count(val))
    return res_val, res_freq

def get_mode(lst):
    lst1 = []
    unique_val, frequency = get_frequency_list(lst)
    max_freq= max(frequency)
    i = 0
    for i in range(len(frequency)):
        if frequency[i] == max_freq:
            break

    return unique_val[i]

# test mode function
# l1 = [1,2,3,4,5,5,4,3,5,8,9,10,4]
# print(f" get mode test exp res 5, val = {get_mode(l1)}")

all_data = []
for file in file_list:
    all_data.append(getDataFromFile(file))

print(len(all_data))

players = ["dhoni", "gambhir", "kapil_dev", "kohli", "rohit", "sachin", "sehwag"]

mean = []
median = []
mode = []
std = []

for player_data in all_data:
    mean.append(np.mean(player_data))
    median.append(np.median(player_data))
    mode.append(get_mode(player_data))
    std.append(np.std(player_data))

print(players, mean, median, mode,std)
```

2. Make a barchart comparing each of the batsmen. That is make 3 plots, one

```python
plt.bar(range(len(players)), mean)
plt.xticks(np.arange(len(players)), players, rotation=45)
plt.ylabel("runs average")
plt.xlabel("player name")
plt.title("Player-wise average runs")
plt.show()
```
