### Given there's a random price change daily in given instrument, if a given amount is invested everyday, what is the average return after n days
define number of days
define daily invetment amount


```python
n = 20 # duration of investment
daily_inv_amt = 7500
min_price = 107.30
max_price = 130.30
delta = max_price - min_price
min_price = min_price - (delta/2)
daily_prices = delta * np.random.rand(n)
for i in range(n):
    daily_prices[i] = daily_prices[i]+min_price
inv_val = []
inv_unit = []
for i in range(n):
    unit_count = daily_inv_amt//daily_prices[i]
    abs_val = unit_count * daily_prices[i]
    bal = daily_inv_amt - abs_val
    inv_unit.append(int(unit_count))
    inv_val.append(abs_val)

total_purchase_units = sum(inv_unit)
total_purchase_amount = sum(inv_val)
total_sell_amount = np.average(daily_prices)*sum(inv_unit)

print(f"min price = {min_price}, max price = {max_price}")
print(daily_prices)
print(np.average(daily_prices))
print(inv_unit)
print(f"No of day = {n}, total buy unit = {total_purchase_units}, buy value= {total_purchase_amount}, average = {total_purchase_amount/total_purchase_units}")
print(f"Sell value= {total_sell_amount}, profit % {(total_sell_amount *100/ total_purchase_amount) - 100}");
# plt.plot(range(1,n+1),daily_prices);
```
