## Q1 Alice Book
1. Consider the Alice in wonderland book (use the text file provided already). Using this data, perform the following tasks:

a. Write code to read the entire novel, remove all punctuation, and replace multiple whitespaces
with a single space. Then count the number of times each word (case insensitive)
occurs in the text. Show a histogram of the top 15 words in the text. Hint: you can
use a collections.Counter or use pd.Series to get the sorted values. (3 marks)

b. Now split the book into the respective chapters. You will need to see the text file and
think about how to do this. Once you have the text of each of the chapters, we will
track the occurrence of 3 characters in the book namely, “Alice”, “White Rabbit”, and
the “Hatter”. Find out in which chapters these characters occur, i.e. for each character
list the chapters that they occur in and also count the number of times they occur in
each chapter. Rank the characters on the number of times they occur in total in the
book. Plot these quantities suitably. (5 marks)

```python
try:
    with open('./Mid Term_data/alice.txt') as f:
        content = f.read()
except FileNotFoundError:
    print("Please check and correct the path of folder. File not found.")

translator = content.maketrans('', '', string.punctuation)
clean_data_1 = content.translate(translator)  # Remove punctuations.

clean_data_2 = ' '.join(clean_data_1.split()) # Replace multiple spaces with single.
word_counts = Counter(clean_data_2.lower().split())
# print(' \n counters \n', word_counts)
```

```python
s1 = pd.Series(word_counts).sort_values(ascending=False)
s1 = s1.head(n=15)
# print('\n',s1[:15].keys);
# print('\n',s1[:15].values);
```

```python
# print(s1.index)
# print(s1.values)
# print(s1.shape)
# plt.hist(outcomes, bins=np.arange(0.5, 7.5, 1), rwidth=0.8, edgecolor='black', color='skyblue')
# plt.hist(s1.values, bins=range(1,16), rwidth=0.8, edgecolor='black', color='skyblue')
plt.bar(range(15), s1.values)
plt.xticks(np.arange(15), s1.index, rotation=45)
plt.ylabel("words")
plt.xlabel("Frequency")
plt.title("Histogram of the top 15 words in the text")
plt.grid()
plt.show()
```

## Q 2 Part B

b. Now split the book into the respective chapters. You will need to see the text file and
think about how to do this. Once you have the text of each of the chapters, we will

list the chapters that they occur in and also count the number of times they occur in

```python
chapter_wise_list = clean_data_2.split("CHAPTER")
# print(type(chapter_wise_list))
chapter_wise_list.pop(0)
# print(len(chapter_wise_list),'\n',chapter_wise_list[0][:20])
# print()
alice = []
white_rabbit = []
hatter = []
for ch_str in chapter_wise_list:
    ch_str = ch_str.lower()
    alice.append(ch_str.count('alice'))
    white_rabbit.append(ch_str.count('white rabbit'))
    hatter.append(ch_str.count('hatter'))
# print(alice)
# print(white_rabbit)
# print(hatter)
df = pd.DataFrame({"Chapter":np.arange(1,13), "Alice":alice, "White Rabbit":white_rabbit, "Hatter": hatter})
print(f"\n word count \n {df}")
print("\n\n Total occurances \n Alice = ",df["Alice"].sum(),
     "\n White Rabbit = ", df["White Rabbit"].sum(),
      "\n Hatter = ", df["Hatter"].sum())

# Bar plot
# df.plot.bar({x=df.columns, y=[df["Alice"].sum(), df["White Rabbit"].sum(), df["Hatter"].sum()]})
aliceData = df["Alice"]
wrData =  df["White Rabbit"]
hatterData = df["Hatter"]
df = pd.DataFrame({'Alice': aliceData, 'White Rabbit': wrData, "Hatter":hatterData}, index=np.arange(1,13))
ax = df.plot.bar()
plt.xlabel("chapter wise word count data")
plt.ylabel("word count")
plt.grid()
```
