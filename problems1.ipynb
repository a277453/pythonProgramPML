 2) Given a number with less than 10 digits, create a list with the digits of the number.

    Input: 45  
    Output: [4,5]

    Input: 62173  
    Output: [6,2,1,7,3]

```python
def getDigiList(val):
    ret = []
    while val != 0:
        ret.insert(0,val%10)
        val = val // 10
    return ret

val = 45
print(getDigiList(val))
print([int(digit) for digit in str(val)]); # List Comprehension
```

    
 3) Sum of Digits in a Number

    Problem:
    Given an integer(less than 10 digits), find the sum of its digits.
    Hint: Use floor division `//` binary operator to get the integer part when you perform a division.
    For example `11//2 ==  5`
    
    Input: 4562  
    Output: 17
 

```python
def getDigitSum(val):
    ret = 0
    val = abs(val)
    while val > 0:
        ret += val%10;
        val = val // 10
    return ret

val = 4562
print(getDigitSum(val))
```

    
 4) Factorial of a Number

    Problem:
    Given an integer n, calculate its factorial (n!) 

    Input: 5  
    Output: 120

```python
def getFact(val):
    ret = 1;
    for i in range(1,val+1):
        ret *= i;
    return ret

val = 5
print(getFact(val));
```

    Problem: Print all numbers from 1 to 100 that are divisible by 4 but not divisible by 8.

  

```python
max = 100

for i in range(1,max+1):
    if not (i % 4) and (i%8):
        print(i, end=' ')
```
