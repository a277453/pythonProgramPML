```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = [1, 2, 3, 4, 5]
values1 = [3,4,6,7,10]
values2 = [16,18,22,24,30]

plt.plot(time, values1)
plt.plot(time, values2)
plt.xlabel('values')
plt.ylabel('reading number');
```

```python
def getVariance(arr):
    total = 0
    for val in arr:
        total += val*val
    print("total ", total, "Average ", np.average(arr) )
    vari = ((total * total) - len(arr)*np.average(arr))/(len(arr) - 1);
    print(' variance ', vari)

getVariance(values1)
getVariance(values2)
```
