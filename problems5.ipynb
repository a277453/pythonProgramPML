# Problem 1:
Simulate  1000 tosses of a biased coin that shows heads (H) with probability 0.7 and tails (T) with probability 0.3.

Plot a histogram of the outcomes.

Compare the observed probabilities with the theoretical probabilities (0.7 and 0.3).

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate the coin tosses
possible_values = ['Heads', 'Tails']
toss_count = 100
samples = np.random.choice(possible_values, toss_count, p=[0.7, 0.3])

# Get the unique values and their counts
u, counts = np.unique(samples, return_counts=True)
print(f"Unique values: {u}")
print(f"Counts: {counts}")

# Plot the bar chart with the correct function
plt.figure(figsize=(6, 5))
plt.bar(u, counts, color=['g','b'], edgecolor='black', alpha=0.5)
plt.yticks(np.arange(0,90,10))
plt.title('Coin Toss Simulation of biased coin \n (heads (H) with probability 0.7 and tails (T) with probability 0.3)')
plt.xlabel('Outcome')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--')
plt.show()
```

# Problem 3:
For n=50 trials with success probability p=0.5:

Plot the binomial probability mass function (pmf).

Compare it with the corresponding normal distribution approximation.

```python
# help from google on what to use and how to plot
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, norm
from math import sqrt

def plot_binomial_and_normal(n, p):
    """
    Plots the binomial probability mass function and its normal approximation.

    Args:
        n (int): The number of trials.
        p (float): The probability of success on each trial.
    """
    # 1. Calculate the parameters for the normal approximation
    mu = n * p  # Mean of the binomial distribution
    sigma = sqrt(n * p * (1 - p))  # Standard deviation

    # 2. Generate x values for plotting
    x_binomial = np.arange(0, n + 1)
    x_normal = np.linspace(0, n, 1000)

    # 3. Calculate binomial probabilities
    pmf = binom.pmf(x_binomial, n, p)

    # 4. Calculate normal distribution values
    pdf = norm.pdf(x_normal, mu, sigma)

    # 5. Create the plot
    plt.figure(figsize=(10, 6))

    # Plot the binomial PMF as a bar chart
    plt.bar(x_binomial, pmf, color='skyblue', edgecolor='black', label='Binomial PMF')

    # Plot the normal approximation as a line
    plt.plot(x_normal, pdf, color='orange', linewidth=2, label='Normal Approximation (PDF)')

    # Add labels, title, and legend
    plt.title(f'Binomial Distribution (n={n}, p={p}) and Normal Approximation', fontsize=16)
    plt.xlabel('Number of Successes (k)', fontsize=12)
    plt.ylabel('Probability', fontsize=12)
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Show the plot
    plt.show()

# Set parameters for the simulation
n_trials = 50
p_success = 0.5

# Call the function to generate the plot and comparison
plot_binomial_and_normal(n_trials, p_success)
```
