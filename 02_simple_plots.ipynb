# Basic plotting
Let us get started with some basic plotting.

Python provides builtin functionality, however plotting and numerical work requires using different packages like numpy and matplotlib.
## Importing packages

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 50)
plt.plot(x, np.sin(x));
```

* np.linspace is a function that creates an array.
* x is an array with 50 floating point numbers from 0 to
* plt.plot plots np.sin(x) versus x with some default line style and color

```python
plt.plot(x, np.sin(x))
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sinusoids')
plt.legend(['sin(x)'])
```

* notice in legend function loc="value" here `loc` is called keyword argument 

```python
plt.plot(x, np.sin(x))
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sinusoids')
plt.legend(['sin(x)'],loc="lower left");
```

```python
# 'b' and 'g' in following calls as third arguments are format strings to specify plot color. They are positional arguments. 
plt.plot(x, np.sin(x), 'b')
# same as plt.plot(x, np.sin(x), color='b')

plt.plot(x, 0.5*np.sin(2*x), 'g', linewidth=2)
#same as plt.plot(x, 0.5*np.sin(2*x), color='g', linewidth=2)

plt.annotate('local maxima', xy=(1.5, 1));
```

Notice that 
* color can be specified as positional arguments as well as keyword argument. 
* that multiple plot statements will place the plots on the same figure

## Exercise: review/homework problem

- Plot $x, -x, \sin(x), x \sin(x)$ in range $-5\pi$ to $5\pi$
- Add a suitable legend
- Annotate the origin (with text 'origin')
- Set axes limits to the range of x

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-5*np.pi, 5*np.pi, 200)
plt.plot(x, -1*x, 'b')
plt.plot(-x,-x, 'b')
plt.plot(x, np.sin(x), "g" )
plt.plot(x, x*np.sin(x), "r", linewidth=2);
plt.legend(['x', '-x', 'sin(x)', 'xsin(x)'],loc="upper right")
plt.text(0.2,0.2, 'origin', color='black')
plt.scatter(0, 0, color='black')
plt.savefig('assignment.png');
```

```python
# tuple unpacking of taking more than one value from function call and store in two specific varia ble
# e.g. var1, var2 = f()
plt.plot(x, np.sin(x), 'b')
xmin, xmax = plt.xlim() # saved returned value in two variables. Tuple unpacking.
ymin, ymax = plt.ylim() # saved returned value in two variables. Tuple unpacking.
print(xmin, xmax, ymin, ymax)
```
